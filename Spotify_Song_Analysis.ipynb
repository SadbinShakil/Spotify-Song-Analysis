## Importing Important Libraries


```python
import numpy             as np
import pandas            as pd
import matplotlib.pyplot as plt
import seaborn           as sns
```

## Viewing the track.csv dataset

```python
df_tracks = pd.read_csv('tracks.csv')
df_tracks.head()
```

## Checking for null values in the dataset

```python
df_tracks.isnull().values.any()
```

```python
pd.isnull(df_tracks).sum()
```

```python
df_tracks.info()
```

## Dropping rows with missing or nan values as they are small in numbers

```python
df_tracks = df_tracks.dropna()
df_tracks.head()
```

```python
pd.isnull(df_tracks).sum()
df_tracks.info()
```

## Find 10 least popular song on spotify

```python
Least_popular_song = df_tracks.sort_values('popularity' , ascending='True')
Least_popular_song.head(10)
```

## Descriptive statistics for each column

```python
df_tracks.describe().transpose()
```

## Finding top 10 popular song with the popularity of greater than or equal 85% 

```python
top_10_song = df_tracks.sort_values(by='popularity', ascending=False)
top_10_song = top_10_song[top_10_song['popularity'] >= 85].head()
top_10_song
 
```

```python
top_10_song
```

## Convert duration miliseconds to seconds

```python
top_10_song['Duration'] = top_10_song['duration_ms'].apply(lambda x: round(x/1000))
top_10_song.drop('duration_ms', inplace=True, axis=1)
```

```python
top_10_song['Duration']
```

## Correlation between variables

```python
Correlation_df = df_tracks.drop(["key","mode","explicit"],axis=1).corr(method = "pearson")
plt.figure(figsize=(14,16))
heatmap = sns.heatmap(Correlation_df, annot = True,fmt = ".1g", vmin = -1, vmax = 1,center = 0, cmap = "inferno", linewidth = 1, linecolor = "Black")
heatmap.set_title("Correlation heatmap between Variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation = 90)
```

## Lets check the correlation with two columns by plotting

```python
Sample_df = df_tracks.sample(int(0.004*len(df_tracks)))
plt.figure(figsize = (10,6))
sns.regplot(data = Sample_df, y = "loudness", x = "energy", color = "c").set(title = "Correlation between energy and loudness")
```

```python
plt.figure(figsize = (10,6))
sns.regplot(data = Sample_df, y = "loudness", x = "popularity", color = "c").set(title = "Correlation between popularity and loudness")
```

```python
plt.figure(figsize = (10,6))
sns.regplot(data = Sample_df, y = "duration_ms", x = "loudness", color = "c").set(title = "Correlation between duration and loudness")
```

## Histogram for no. of songs per year...

```python
df_tracks['dates'] = df_tracks.index.get_level_values('release_date')
df_tracks.dates = pd.to_datetime(df_tracks.dates)
years = df_tracks.dates.dt.year
sns.displot(years,discrete=True,aspect=2,height=5,kind="hist").set(title="Number of songs per year")
```

## Duration of songs over years

```python
total_duration = df_tracks.duration_ms
fig_dims = (18,7)
fig, ax = plt.subplots(figsize=fig_dims)
fig = sns.barplot(x = years, y = total_duration, ax = ax, errwidth = False).set(title= "Year vs duration")
plt.xticks(rotation=90)
```

```python
total_duration = df_tracks.duration_ms
sns.set_style(style="whitegrid")
fig_dims = (10,5)
fig, ax = plt.subplots(figsize=fig_dims)
fig = sns.lineplot(x = years, y = total_duration, ax = ax).set(title= "Year vs duration")
plt.xticks(rotation=60)
```
